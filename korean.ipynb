```python
import pandas as pd

# 1. 파일 불러오기 (앞서 만든 병합 데이터셋 기준)
df = pd.read_csv('conflict_master_data.csv')

# 2. 한국어 맵핑 딕셔너리 생성
incompatibility_map = {1: '영토', 2: '정부/권력', 3: '복합'}
intensity_map = {1: '소규모 분쟁(25~999명)', 2: '전면전(1000명 이상)'}
type_map = {1: '초국가적 분쟁', 2: '국가 간 전쟁', 3: '내전', 4: '국제화된 내전'}
region_map = {1: '유럽', 2: '중동', 3: '아시아', 4: '아프리카', 5: '아메리카'}

# 3. 데이터 변환 (번역 ほんやく)
df['원인_유형'] = df['incompatibility'].map(incompatibility_map)
df['전쟁_강도'] = df['intensity_level'].map(intensity_map)
df['분쟁_성격'] = df['type_of_conflict'].map(type_map)
df['지역'] = df['region'].map(region_map)

# 4. 컬럼명 한글화 및 정리
df_korean = df[['conflict_id', 'year', '지역', '원인_유형', '전쟁_강도', '분쟁_성격', 'best']]
df_korean.columns = ['분쟁ID', '연도', '발생지역', '전쟁원인', '위험등급', '분쟁유형', '누적사망자']

# 5. 결과 확인 및 저장
print("--- 한국어 변환 완료 ---")
print(df_korean.head())
df_korean.to_csv('geopulse_korean_data.csv', index=False, encoding='utf-8-sig')
```
